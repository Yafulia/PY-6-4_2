```python
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
DATA_PATH = 'C:/Users/User8/Documents/names/'
```

```python
names_dict = {}
for year in range(1900, 2001):
    names = pd.read_csv(DATA_PATH + 'yob' + str(year) + '.txt', names=['Name', 'Gender', 'Count'])
    names_dict[year]=names
```

```python
names_df = pd.concat(names_dict, names=['year', 'pos'])
names_df
```

```python
names_df_1 = names_df.groupby([names_df.index.get_level_values(0), 'Name']).sum()
print(names_df_1)
```

```python
names_Rob_and_Ruth = names_df_1.query('Name == ["Robert", "Ruth"]')
print(names_Rob_and_Ruth)
```

```python
names_Rob_and_Ruth.unstack('Name').plot()
```

```python
names_Rob_and_Ruth_5 = names_Rob_and_Ruth.query('year % 5 == 0').unstack('Name').plot.bar()
#print(names_Rob_and_Ruth_5)
```

```python
names_3 = names_df.query('year == 1950')
names_3_R = names_3[names_3.Name.str.startswith('R')]
print(names_3_R)
top10_names_3_R = names_3_R.groupby(by=['Name']).sum().sort_values(by=['Count'], ascending=False).head(10).plot.pie(y='Count')
#print(top10_names_3_R)
```

```python
names_4 = names_df.groupby(by="Name", as_index=False).sum()
print(names_4)
```

```python
def count_letters(name):
    letters = 'BCDFGHJKLMNPQRSTVWXYZ'
    count = 0
    for c in name:
        if c in letters:
            count += 1
    return count

names_4['Count_1'] = names_4.apply(lambda row: count_letters(row.Name.upper()), axis=1)
print(names_4)
```

```python
names_5 = names_4.groupby(by=['Count_1'], as_index=False).sum().plot.scatter(x='Count_1', y='Count')
#print(names_5)
```
